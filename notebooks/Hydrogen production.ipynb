## Biomass and Hydrogen flux

```python
with auto:
    print(model_names[0])
    # Set objective to biomass production
    auto.objective = auto.reactions.Biomass_Chlamy_auto
    print(auto.objective)
    solution = auto.optimize()
    print("Biomass flux:\t\t", solution.fluxes.Biomass_Chlamy_auto)
    print("H2 Exchange flux:\t", solution.fluxes.EX_h2_e)
    # Set objective to H2 production
    auto.objective = auto.reactions.get_by_id('EX_h2_e')
    print("\n", auto.objective)
    solution = auto.optimize()
    print("Biomass flux:\t\t", solution.fluxes.Biomass_Chlamy_auto)
    print("H2 Exchange flux:\t", solution.fluxes.EX_h2_e)
    
    print()
with mixo:
    print(model_names[1])
    # Set objective to biomass production
    mixo.objective = mixo.reactions.Biomass_Chlamy_mixo
    print(mixo.objective)
    solution = mixo.optimize()
    print("Biomass flux:\t\t", solution.fluxes.Biomass_Chlamy_mixo)
    print("H2 Exchange flux:\t", solution.fluxes.EX_h2_e)
    # Set objective to H2 production    
    mixo.objective = mixo.reactions.get_by_id('EX_h2_e')
    print("\n", mixo.objective)
    solution = mixo.optimize()
    print("Biomass flux:\t\t", solution.fluxes.Biomass_Chlamy_mixo)
    print("H2 Exchange flux:\t", solution.fluxes.EX_h2_e)

    print()
with hetero:
    print(model_names[2])
    # Set objective to biomass production
    hetero.objective = hetero.reactions.Biomass_Chlamy_hetero
    print(hetero.objective)
    solution = hetero.optimize()
    print("Biomass flux:\t\t", solution.fluxes.Biomass_Chlamy_hetero)
    print("H2 Exchange flux:\t", solution.fluxes.EX_h2_e)
    # Set objective to H2 production
    hetero.objective = hetero.reactions.get_by_id('EX_h2_e')
    print("\n", hetero.objective)
    solution = hetero.optimize()
    print("Biomass flux:\t\t", solution.fluxes.Biomass_Chlamy_hetero)
    print("H2 Exchange flux:\t", solution.fluxes.EX_h2_e)
```

When setting the objective to maximize biomass flux, the model yields the highest hydrogen production in autotrophic conditions. In heterotrophic conditions no hydrogen at all is produced, while in mixotrophic a very small amount is released. 

To see the theoretical maximal capacity to produce hydrogen, the objective was also set to maximize hydrogen export. Interestingly, here the model yields the highest hydrogen export in mixotrophic conditions.

## Flux Scanning based on Enforced Objective Flux

In order to increase the production hydrogen, possible gene amplification targets can be identified using flux scanning based on enforced objective flux (FSEOF). "In particular, FSEOF is a method that first scans and searches for variations in the metabolic fluxes in response to the enforced fluxes directed towards a target product. Reactions were then selected as amplification targets, the flux values of which increased in accordance with the enforced fluxes toward the production of a target chemical."

Using the autotrophic model, objective fluxes of hydrogen export are enforced and the changes of metabolic fluxes in response to that are analysed. 



```python
from cameo.strain_design.deterministic.flux_variability_based import FSEOF
```

```python
fseof = FSEOF(auto)
```

```python
result = fseof.run(target=auto.reactions.EX_h2_e)
```

```python
result.data_frame
```

```python
from matplotlib import pyplot as plt
result.data_frame.T.plot()
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

## Differential flux variability analysis

Compares flux ranges of a reference model to a set of models that have been parameterized to lie on a grid of evenly spaced points in the n-dimensional production envelope (n being the number of reaction bounds to be varied).

```python
auto.objective = auto.reactions.Biomass_Chlamy_auto
```

```python
from cameo.strain_design.deterministic import DifferentialFVA
```

```python
# Reference state model
reference_model = auto.copy()
biomass_rxn = reference_model.reactions.Biomass_Chlamy_auto
biomass_rxn.lower_bound = 0.
target = reference_model.metabolites.h2_e
```

```python
# Set up the differential flux variability analysis strain design method
diff_fva = DifferentialFVA(design_space_model=auto,
                           reference_model=reference_model,
                           objective=target,
                           variables=[biomass_rxn],
                           normalize_ranges_by=biomass_rxn,
                           points=10)
```

```python
# Run differential flux variability analysis (only on the surface of the production envelope)
%time result = diff_fva.run(surface_only=True)
result.plot()
```
